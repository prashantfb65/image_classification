```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
img = cv2.imread('data/00-puppy.jpg')
```

```python
plt.imshow(img)
# Matplotlib --> RGB
# OpenCV --> GBR
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('data/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
```

#### Resizing of image

```python

print(fix_img.shape)
new_img = cv2.resize(fix_img, (1000, 400))
```

```python
plt.imshow(new_img)
```

```python
w_ratio = 0.8
h_ratio = 0.2
new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
```

```python
plt.imshow(new_img)
```

#### Flipping of images

```python
new_img = cv2.flip(fix_img, 1)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, -1)
plt.imshow(new_img)
```
