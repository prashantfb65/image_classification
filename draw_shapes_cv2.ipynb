```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blnk_img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
blnk_img.shape
```

```python
plt.imshow(blnk_img)
```

```python
cv2.rectangle(blnk_img, pt1=(384,10), pt2=(490, 150), \
              color=(0,255,0), thickness=10)
plt.imshow(blnk_img)
```

```python
cv2.rectangle(blnk_img, pt1=(200, 200), pt2=(300,300), \
              color=(0,255,255), thickness=10)
plt.imshow(blnk_img)
```

```python
cv2.circle(blnk_img, center=(100, 100), radius=50, \
              color=(255,0, 5), thickness=10)
plt.imshow(blnk_img)
```

```python
cv2.circle(blnk_img, center=(400, 400), radius=50, \
              color=(255,0, 255), thickness=-1)
plt.imshow(blnk_img)
```

```python
cv2.line(blnk_img, pt1=(100, 200), pt2=(200,300), \
              color=(0,150,255), thickness=10)
plt.imshow(blnk_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blnk_img, text='Hello', org=(10, 500),
           fontFace=font, fontScale=4, color=(255,255,255),
           thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blnk_img)
```
